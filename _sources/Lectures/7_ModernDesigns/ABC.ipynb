The limitations in thrust-producing abilities of the rotor on the retreating sides, due to low dynamic pressure, are a significant impediment to exploiting the full aerodynamic lifting characteristics of the blades. This is apparent from the figure above.

```{toggle}
The advancing blade tips would start experiencing high drag as Mach divergence number (~0.9) is achieved. So to fully exploit the advantages of ABC, the rotors need to be slowed down. In the original ABC concept demonstrator (XH-49A) the tip speed was reduced from 198 m/s to 137 m/s 
```

## Sikorsky XH-49A

Coaxial rotors
Thrust-compounding with turbojets

## How it started

<div class="image-grid">
    <img  src="../../assets/P152_1.png" class="big"/>
    <img  src="../../assets/P152_2.png" class="big"/>
    <img  src="../../assets/P152_3.png" class="big" />

   
</div>

<div class="image-grid">
  <p style="font-size:55%; color:gray; text-align:left;"> ABC demonstrator <a href="https://vtol.org/store/product/the-abc-rotor-a-historical-perspective-4111.cfm"> [source]</a></p>   
  <p style="font-size:55%; color:gray; text-align:left;"> ABC concept aircraft <a href="https://vtol.org/store/product/the-abc-rotor-a-historical-perspective-4111.cfm"> [source]</a></p>
  <p style="font-size:55%; color:gray; text-align:left;"> ABC concept armoured aircraft <a href="https://vtol.org/store/product/the-abc-rotor-a-historical-perspective-4111.cfm"> [source]</a></p>
</div>

## How it is going

<div class="image-grid">
    <img  src="../../assets/X2.jpeg" class="big"/>
    <img  src="../../assets/Raider.jpeg" class="big"/>
    <img  src="../../assets/Defiant.jpg" class="big" />

   
</div>

<div class="image-grid">
  <p style="font-size:55%; color:gray; text-align:left;"> X2 <a href="https://www.wired.com/2010/08/sikorsky-x2-breaks-helicopter-speed-record/"> [source]</a></p>   
  <p style="font-size:55%; color:gray; text-align:left;"> ABC concept aircraft <a href="https://en.wikipedia.org/wiki/Sikorsky_S-97_Raider#/media/File:Sikorsky_S-97_Raider_in_flight.jpg"> [source]</a></p>
  <p style="font-size:55%; color:gray; text-align:left;"> ABC concept armoured aircraft <a href="https://en.wikipedia.org/wiki/Sikorsky%E2%80%93Boeing_SB-1_Defiant#/media/File:Sikorsky%E2%80%93Boeing_SB-1_Defiant_(cropped).jpg"> [source]</a></p>
</div>

X2 set a speed record of 468 km/hr {cite}`Add48`.

## Advantages

- permits large variation in helicopter CG

- (B28) retreating side doesn't need to operate at high CL anymore so stall can be avoided there

## So, are all future helicopters going to be ABC designs?
