## Induced power

The induced velocity in forward flight is stated below. It is to be assumed as agiven. It can be derived using the same principles that the expression for hover and climb cases were obtained, and is again based on the uniform inflow velocity assumption. 

$$v_i = \frac{T}{2\rho A U}$$

where $U = \left | \bar{V_{\infty}} + \bar{v_i} \right |$ is the vector sum of the oncoming flow velocity and the induced flow velocity. This result is consistent with the expressions obtained for hover and axial flight in that it reduces to the corresponding expressions when appropriate substitutions are made for U and this is shown below. 

In the case of fast forward flight, $V_{\infty} \text{ or } V_f>>v_i$ and therefore the induced velocity is alternately stated as follows-

$$v_i = \frac{T}{2\rho A U} \approx \frac{T}{2\rho A V_{\infty}}$$

### Hover

<div style="float:right; width:43%;" >
    <img src="../../assets/HT_glauert_hover.png" class="big" />
    <p style="font-size:16px; color:gray; text-align:center;" > Hover <a href=""> </a></p>
</div>

\begin{equation}
\begin{aligned}
v_{i} &=\frac{T}{2 \rho A U} \\
&=\frac{T}{2 \rho A \sqrt{\left(V_{\infty} \cos \alpha\right)^{2}+\left(-V_{\infty} \sin \alpha+v_{i}\right)^{2}}} \\
&=\frac{T}{2 \rho A \sqrt{V_{\infty}^{2} \cos ^{2} \alpha+V_{\infty}^{2} \sin ^{2} \alpha-2 V_{\infty} \sin \alpha v_{i}+v_{i}^{2}}} \\
&=\frac{T}{2 \rho A \sqrt{V_{\infty}^{2}-2 V_{\infty} \sin \alpha v_{i}+v_{i}^{2}}} \\
&=\frac{T}{2 \rho A \sqrt{v_{i}^{2}}} \\
&=\frac{T}{2 \rho A v_{i}} \\
v_{i}^{2} &=\frac{T}{2 \rho A} \\
v_{i} &=\sqrt{\frac{T}{2 \rho A}}
\end{aligned}
\end{equation}

### Vertical climb

<div style="float:right; width:43%;" >
    <img src="../../assets/HT_glauert_climb.png" class="big" />
    <p style="font-size:16px; color:gray; text-align:center;" > Vertical climb <a href=""> </a></p>
</div>

\begin{equation}
\begin{aligned}
v_{i} &=\frac{T}{2 \rho A \sqrt{V_{\infty}^{2}-2 V_{\infty} \sin \alpha v_{i}+v_{i}^{2}}} \\
&=\frac{T}{2 \rho A \sqrt{V_{\infty}^{2}+2 V_{\infty} v_{i}+v_{i}^{2}}} \\
&=\frac{T}{2 \rho A\left(V_{\infty}+v_{i}\right)} \\
&=-\frac{V_{\infty}}{2}+\sqrt{\frac{T}{2 \rho A}+\frac{V_{\infty}^{2}}{4}} \\
&=-\frac{V_{\infty}}{2}+\sqrt{v_{h}^{2}+\frac{V_{\infty}^{2}}{4}}
\end{aligned}
\end{equation}

### Low forward speed 

<div style="float:right; width:43%;" >
    <img src="../../assets/HT_glauert_low_speed.png" class="big" />
    <p style="font-size:16px; color:gray; text-align:center;" > Low speed forward flight <a href=""> </a></p>
</div>

\begin{equation}
\begin{aligned}
v_{i} &=\frac{T}{2 \rho A \sqrt{V_{\infty}^{2}-2 V_{\infty} \sin \alpha v_{i}+v_{i}^{2}}} \\
& \approx \frac{T}{2 \rho A \sqrt{V_{\infty}^{2}+v_{i}^{2}}} \\
\Rightarrow v_i &=\sqrt{-\frac{V_{\infty}^{2}}{2}+\sqrt{\left(\frac{T}{2 \rho A}\right)^{2}+\frac{V_{\infty}^{4}}{4}}} \\
&=\sqrt{-\frac{V_{\infty}^{2}}{2}+\sqrt{v_{h}^{4}+\frac{V_{\infty}^{4}}{4}}}
\end{aligned}
\end{equation}

The low forward speed assumption of $\alpha\approx 0$ is generally valid for advance ratios $\mu<0.15$, where $\mu = \frac{V_{\infty}}{\Omega R}$.

</div>

&nbsp;
&nbsp;
&nbsp;

\begin{equation}
\begin{aligned}
v_{i} &=\frac{T}{2 \rho A \sqrt{V_{\infty}^{2}-2 V_{\infty} \sin \alpha v_{i}+v_{i}^{2}}} \\
& \approx \frac{T}{2 \rho A \sqrt{V_{\infty}^{2}}} \\
&=\frac{T}{2 \rho A V_{\infty}}
\end{aligned}
\end{equation}

The inflow variation with forward speed can then be obtained by *stitching-up* the results in the different domains. Once the induced velocity is known, the induced power can be simply obtained as $Tv_i$. As was mentioned earlier for hover, large mass flow through the rotor disk is key to thrust efficiency of a rotor. When a rotor is tilted for forward flight, a component of the oncoming flow velocity goes through the rotor disk. This increases the effective mass flow rate of air through the disk much like in the rotor climb scenario. Consequently, a smaller increase in velocity needs to be imparted to the flow to get the required momentum change and leads to the reduction in $v_i$ obtained.

## Aside: rotor aerodynamics in descent flight

Descent flight in rotorcraft is usually undertaken at very low speeds. Fast descent speeds can lead to large flow recirculation and turbulence in the flow, leading to high vibrations and loss of control. There are three distinct states that exist with increasing descent flight velocities of a rotor - 

Vortex ring state 
- $V<0$ but $V+v_i>0$

Turbulent wake state
- $V<0$ and $V+v_i<0$, but $V+2v_i>0$

Windmill brake state
- $V<0$, $V+v_i<0$, $V+2v_i<0$

## Rotor aerodynamics assymetry

As soon as a rotor starts moving edgewise, a lateral assymetry is introduced with regard to the total in-plane velocity. This assymetry is most prominent at the azimuth locations $\psi=90$ and $\psi=270$ and decreases progressively to no assymetry at azimuth locations $\psi=0$ and $\psi=180$.

## Reverse flow

Along with the in-plane flow assymetry in the lateral direction, a reverse flow region occurs in forward flight. This is of concern because the blade sections within this region effectively have an angle of attack close to 180 and are no longer aerodynamically efficient. There is no minimum threshold speed for the reverse flow region to occur but it is not of major concern for low speed flight since rotors generally have a root cutout region of about 20%. Based on the equations given below, this means that only for $\mu > 0.2$ do blade airfoil sections start experiencing reverse flow.

$$ V_{T}=\Omega \cdot y+V_{F} \cdot \sin \Psi $$


$$ V_{T}=\Omega \cdot y+V_{F} \cdot \sin \Psi=0 $$


$$ y=\frac{-V_{F} \cdot \sin \Psi}{\Omega}=-\frac{V_{F}}{\Omega R} \sin \Psi \cdot R=-\mu \cdot \sin \Psi \cdot R$$

## Profile power

The drag due to fluid viscosity and due to pressure distribution over the airfoil surface manifests itself within the airfoil drag coefficient. The power required by the rotor to overcome these drag forces can be evaluated by integrating the power required to overcome drag at each blade section and averaged over the azimuth.

    <img  src="../../assets/HT_profile_power.png" class="big"/>
</div>

<div class="image-grid">
    <p  style="font-size:16px; color:gray; text-align:center;"> Blade section drag due to in-plane velocity </p>
    <p  style="font-size:16px; color:gray; text-align:center;"> Rotor in forward flight </p>
</div>



\begin{equation}
\begin{array}{l}
\displaystyle P_{0}=\frac{N_{b}}{2 \pi} \int_{0}^{2 \pi } \int_{0}^{R} P_{0}^{\prime} d y d \Psi \\
\displaystyle P_{0}=\frac{N_{b}}{2 \pi} \int_{0}^{2 \pi } \int_{0}^{R} D^{\prime}.V_T d y d \Psi \\
\displaystyle P_{0}=\frac{N_{b}}{2 \pi} \int_{0}^{2 \pi} \int_{0}^{R} \frac{1}{2} \rho C_{d 0} c V_{T}^{2} \cdot\left|V_{T}\right| d y d \Psi \\
\displaystyle P_{0}=\frac{N_{b}}{4 \pi} \rho C_{d 0} c \int_{0}^{2 \pi} \int_{0}^{R} V_{T}^{2} \cdot\left|V_{T}\right| d y d \Psi \\
\displaystyle P_{0}=\frac{N_{b}}{4 \pi} \rho C_{d 0} c \cdot\left(\int_{0}^{2 \pi} \int_{0}^{R} V_{T}^{3} d y d \Psi-2 \cdot \int_{\pi}^{2 \pi} \int_{0}^{\frac{-V_{F}sin\Psi}{\Omega}} V_{T}^{3} d y d \Psi\right) \\
\displaystyle P_{0}=\frac{N_{b}}{4 \pi} \rho C_{d 0} c\left(\int_{0}^{2 \pi} \int_{0}^{R}\left(\Omega \cdot y+V_{F} \cdot \sin \Psi\right)^{3} d y d \Psi-2 \cdot \int_{\pi}^{2 \pi} \int_{0}^{\frac{-V_{F} \cdot \sin \Psi}{\Omega}} \left(\Omega \cdot y+V_{F} \cdot \sin \Psi\right)^{3} d y d \Psi\right) \\
\displaystyle P_{0}=\frac{N_{b}}{4 \pi} \rho C_{d 0} c\left(\frac{\pi}{2} \Omega^{3} R^{4}+\frac{3 \pi}{2} \Omega R^{2} V_{F}^{2}+2 \cdot \frac{3 \pi V_{F}^{4}}{32 \Omega}\right) \\
\displaystyle P_{0}=\frac{N_{b}}{4 \pi} \rho C_{d 0} c\left(\frac{\pi}{2} \Omega^{3} R^{4}+\frac{3 \pi}{2} \Omega R^{2} V_{F}^{2}+\frac{3 \pi V_{F}^{4}}{16 \Omega}\right) \\
\displaystyle P_{0}=\frac{N_{b}}{8} \rho C_{d 0} c \Omega^{3} R^{4}\left(1+3 \frac{V_{F}^{2}}{\Omega^{2} R^{2}}+\frac{3 V_{F}^{4}}{8 \Omega^{4} R^{4}}\right) \\
\displaystyle P_{0}=\frac{N_{b}}{8} \rho C_{d 0} c \Omega^{3} R^{4}\left(1+3 \mu^{2}+\frac{3}{8} \mu^{4}\right) \\
\displaystyle P_{0}=\frac{1}{8} \frac{c N_{b}}{\pi R} \pi \rho C_{d 0} \Omega^{3} R^{5}\left(1+3 \mu^{2}+\frac{3}{8} \mu^{4}\right) \\
\displaystyle P_{0}=\frac{1}{8} \rho \sigma A C_{d 0} V_{TIP}^{3}\left(1+3 \mu^{2}+\frac{3}{8} \mu^{4}\right)
\end{array}
\end{equation}

## Parasite power

The power expended in order to overcome the fuselage drag of a helicopter is refered to as the parasite power. It is simply the parasite drag times the forward speed where the parasite drag itself is usually stated as an equivalent flat plate area $f$. Since fuselage attitude largely remains the same at any speed, $f$ can be assumed a constant.

$$P_p = D_p*V_{\infty} = \frac{1}{2}\rho V_f^3 f$$

## Power bucket curve

The consumed power breakdown of the entire helicopter is shown below. The power required by the tail rotor $P_{TR}$ which can be fairly approximated as a fixed percentage (about 10%) of the total power required by the main rotor [^TailRotor]. $P_a$ here refers to accesories power for onboard equipment and $P_{tl}$ refers to additional power required to overcome transmission losses. These three power components are small in comparision of the main rotor power $P_i+P_0$ and the helicopter parasite power $P_p$. The characteristic 'bucket' curve is a byproduct of the interplay between the latter 3 power components.

[^TailRotor]: Alternately, the tail rotor power can be also obtained from first principles using BET or momentum theory.
